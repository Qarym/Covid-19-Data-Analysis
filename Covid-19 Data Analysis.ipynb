## Visualize population to test done ratio

```python
pop_test_ratio=world_data['Population']/world_data['TotalTests']
```

```python
pop_test_ratio=world_data['Population']/world_data['TotalTests'].iloc[0:20]
```

```python
fig=px.bar(world_data.iloc[0:20],x='Country/Region',y=pop_test_ratio[0:20],color='Country/Region',title='Pop to Test Done Ratio')
fig.show()
```

## Worst 20 countries having maximum confirmned cases.

```python
fig=px.bar(world_data.iloc[0:20],y='Country/Region',x='TotalCases',color='TotalCases',text='TotalCases')
fig.update_layout(template='plotly_dark',title_text='Top 20 Worst Affected Countries')
fig.show()
```

## Worst 20 countries having maximum TotalDeaths

```python
world_data.sort_values(by='TotalDeaths',ascending=False)
```

```python
fig=px.bar(world_data.sort_values(by='TotalDeaths',ascending=False)[0:20],y='Country/Region',x='TotalDeaths',color='TotalDeaths',text='TotalDeaths')
fig.update_layout(template='plotly_dark',title_text='Worst 20 According to TotalDeaths Countries')
fig.show()
```

## Worst 20 countries having Active cases

```python
fig=px.bar(world_data.sort_values(by='ActiveCases',ascending=False)[0:20],y='Country/Region',x='ActiveCases',color='ActiveCases',text='ActiveCases')
fig.update_layout(template='plotly_dark',title_text='Worst 20 According to ActiveCases Countries')
fig.show()
```

## 20 countries having maximum Recovered Cases

```python
fig=px.bar(world_data.sort_values(by='TotalRecovered',ascending=False)[0:20],y='Country/Region',x='TotalRecovered',color='TotalRecovered',text='TotalRecovered')
fig.update_layout(template='plotly_dark',title_text='20 countries having highest Recovered Cases')
fig.show()
```
